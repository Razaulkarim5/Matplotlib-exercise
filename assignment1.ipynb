```python
#Code here
fig = plt.figure()

# Add an axis to the figure canvas
ax = fig.add_axes([0, 0, 1, 1])  # [left, bottom, width, height]

# Plot the data
ax.plot(x, y, color='green', label='y = x*2')

# Set labels and title
ax.set_title('Plot_title')
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')


ax.legend(loc=4)

# Show the plot
plt.show()
```

**2. Instructions**<br>
* ** Create a figure object**
* ** Put two axes on the figure at [0,0,1,1] and [0.1,0.6,0.2,0.2]**
* ** Plot data (x,z) on main axes**
* ** Plot data (x,z) on the small axes and set its x limit b/w 1, 2, y limit 0, 5**
* ** Replicate the figure to axes, titles, legends etc**

```python

x = np.linspace(0, 10,100)
y_main = x ** 2
y_inset = x ** 2

# Creating an empty canvas
fig3 = plt.figure(figsize=(4,3))

# Creating main axes - values b/w 0 & 1
axes_main = fig3.add_axes([0, 0, 1, 1])

# Creating inset axes - values b/w 0 & 1
axes_inset = fig3.add_axes([0.2, 0.6, 0.25, 0.25]) 

# Main axes plot
axes_main.plot(x, y_main, 'r', label='y_main = x_main**2')
axes_main.set_xlabel('x_main')
axes_main.set_ylabel('y_main')
axes_main.set_title('Title_main')
axes_main.set_ylim(0, 80) 
axes_main.set_xlim(0, 10) 
axes_main.legend(loc=4)

# Inset axes plot
axes_inset.plot(x, y_inset, 'b')
axes_inset.set_xlim(1, 2)
axes_inset.set_ylim(0, 5)
axes_inset.set_xlabel('x_inset')
axes_inset.set_ylabel('y_inset')
axes_inset.set_title('inset')

# Save the figure
fig3.savefig("updated_plot.jpg", dpi=200, bbox_inches='tight')

# Show the plot
plt.show()
```

**5. plot `(x,y)`, `(x,z)` and `(x,y)` on the above empty canvas to replicate the plot given below.**

```python
#Code here


x = np.arange(0, 10, 1)
y1 = x
y2 = x ** 2
y3 = x * 2

# Create three subplots in a vertical layout
fig, axes = plt.subplots(3, 1, figsize=(4, 4))  # 3 rows, 1 column

# Plot on the first subplot
axes[0].plot(x, y1, 'b--')  # Blue dashed line
axes[0].set_title('y = x')

# Plot on the second subplot
axes[1].plot(x, y2, 'r-')  # Red solid line
axes[1].set_title('y = x^2')

# Plot on the third subplot
axes[2].plot(x, y3, 'g-')  # Green solid line
axes[2].set_title('y = 2x')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```

**6. Resize the above plots to `figsize = (3,4)`**

```python
#Code here




# Create three subplots in a vertical layout
fig, axes = plt.subplots(3, 1, figsize=(2, 3))  # Adjusting figsize for a compact layout

# Plot on the first subplot
axes[0].plot(x, y1, 'b--')  # Blue dashed line
axes[0].set_title('y = x')

# Plot on the second subplot
axes[1].plot(x, y2, 'r-')  # Red solid line
axes[1].set_title('y = x^2')

# Plot on the third subplot
axes[2].plot(x, y3, 'g-')  # Green solid line
axes[2].set_title('y = 2x')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show the plot
plt.show()
```
